# Body-site dataset (no disponemos de acceso a datos)

## 1. Reading fasta files

```python
### getting a list of fasta files need to be read from directory
files=FileUtility.recursive_glob('../data_config/bodysites/','*.txt')
list_of_files=[]
for file in files:
    list_of_files+=FileUtility.load_list(file)    
list_of_files=[x+'.fsa' for x in list_of_files]
#print(list_of_files)

### reading fasta files
# return: list of fasta files, and a dic to map file to index
# load only only_files
fasta_files, mapping = FileUtility.read_fasta_directory('../../../datasets/deepbio/microbiome/hmb_data/','fsa',only_files=list_of_files)
```

## 2. Creating bootstrapping object

```python
## create a bootstrapping object with the parameter explaind in the paper
BS=BootStrapping(fasta_files,'output_dir/', seqtype='fasta' ,sampling_sizes=[10, 20, 50, 100, 200, 500, 1000, 2000, 5000, 10000],
                 n_resamples=10, M=1)
```

## 3. Bootstrapping for different k-mers

```python
## add k-mer samplings
for k in [3,4,5,6,7,8]:
    print(k,'-mer bootstrapping..')
    BS.add_kmer_sampling(k)
```

## 4. Plotting

```python
import matplotlib.pyplot as plt
%pylab inline
%matplotlib inline 

BS.plotting('results_bootstrapping_body-site','body-site')
```

# 5. Saving

```python
BS.save_me('body-site_bootstrapping')
```

### Como no tenemos el precalculado, lo calculamos como en el ejemplo anterior

```python
# LER OS FASTAQ
# devolve unha lista de arquivos fastaq (fasta_files) e un diccionario cos arquivos xunto co seu nmero de ndice (mapping);
#  dicir, a orde na que se leron. Por exemplo, o primeiro arquivo fastaq ERR1368879 vemos que ten o ndice 0

fasta_files, mapping = FileUtility.read_fasta_directory('../../crohns_disease/sra_download/samples_fastq/', 'fastq')
```

```python
# CONSTRUIR OBXECTO PARA O BOOTSTRAPPING
'''
input_dir: input directory or a list of files
output_dir: a directory to generate the output files
sampling_sizes: a list of sampling sizes (N's)
n_resamples: number of resamples from each file (N_R = 10)
seqtype: file suffixes fastq or fasta etc.
M: number of files from the directory to make samplings from (M)
'''

BS=BootStrapping(fasta_files,'../../crohns_disease/results/bootstrapping/', seqtype='fastq',
                 sampling_sizes=[10, 20, 50, 100, 200, 500, 1000, 2000, 5000, 10000], n_resamples=10, M=10)
```

```python
## add k-mer samplings
for k in [3,4,5,6,7,8]:
    print(k,'-mer bootstrapping..')
    BS.add_kmer_sampling(k)
```

```python
import matplotlib.pyplot as plt
%pylab inline
%matplotlib inline 

BS.plotting('results_bootstrapping_crohns_disease','crohns_disease')
```

```python
# GARDAR RESULTADOS EN ARQUIVO 'bootstrapping_crohns_disease_prediction' NA RUTA QUE SE LLE INDICOU AO DEFINIR A INSTANCIA BS
BS.save_me('bootstrapping_crohns_disease_prediction')
```
